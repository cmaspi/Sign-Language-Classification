````python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import seaborn as sns
import sys
import os
from tqdm import tqdm
import dataset
````

````python
Images, labels = dataset.get_random_uniform(150)
````

````python
sns.set()
fig, ax = plt.subplots(2,2, figsize=(10, 8))
for i, idx in enumerate(np.random.choice(labels.size, 4, replace=False)):
    t = ax[i//2][i%2].imshow(Images[idx], cmap='gray')
    ax[i//2][i%2].set_title(labels[idx])
    ax[i//2][i%2].axis('off')
    plt.colorbar(t)
plt.show()
sns.reset_orig()
````

````python
sns.set()
fig, ax = plt.subplots(4, 5, figsize=(20,16))
for i in range(20):
    ax[i//5][i%5].imshow(Images[150*i], cmap='gray')
    ax[i//5][i%5].axis('off')
plt.tight_layout()
plt.show()
sns.reset_orig()
    
````
